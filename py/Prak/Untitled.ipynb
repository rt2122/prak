```python
lens = 132
```

```python
for i in range(lens):
    name = data1["Name"][i]
    res = 0
    for j in range(lens - 1):
        if data2["Name"][j] == name:
            res = j
            break
    if res == 0:
        print(name)
    data1["Rating"][i] = data2["ПЛАН. ОЦЕНКА"][res]
```
