```python
df = pd.read_csv("train.csv")
```

```python
signs = ("линии", "возраст", "поведение_30-59_дней", 
         "Debt_Ratio", "доход", "число_кредитов", "поведение_90_дней", "недвижимость", 
         "поведение_60-89_дней", "семья")
target_sign = "плохой_клиент"
target = df.pop(target_sign)
```

```python
df.fillna(-11, inplace=True)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
svm.fit(df, target)
test = pd.read_csv("test.csv")
test.fillna(-11, inplace=True)
ans = svm.predict(test)
```

```python
#c нуля
train = pd.read_csv("train.csv")
train = sk.utils.shuffle(train)
train.shape
```

```python
train.head(10)
```

```python
train.fillna(-11, inplace=True)
```

```python
sep = 80000
train4 = train[:sep]
test4 = train[sep:]
```

```python
train4.head(4)
```

```python
test4.head(4)
```

```python
test_target = test4.pop("плохой_клиент")
train_target = train4.pop("плохой_клиент")
svm = SVC(kernel='rbf')
svm.fit(train4, train_target)
ans = svm.predict(test4)
```

```python
num = 0
test_target = list(test_target)
for i in range(len(ans)):
    if int(ans[i]) == int(test_target[i]):
        num += 1
print(num / len(ans))
```
